# Introduction

Neural networks are computational models that use layers of weighted connections and biases to transform raw input data into meaningful patterns. Each layer applies a weighted sum and a non-linear activation, allowing the network to capture complex, non-linear relationships. After passing through several layers, the model produces an output, such as a class prediction.

In this project, we focus on Convolutional Neural Networks (CNNs), which are specifically designed for grid like data such as images. CNNs use convolutional layers to detect local features, pooling layers to reduce spatial dimensions, and fully connected layers to perform classification. While a CNN may achieve high accuracy, it is also important to understand why it makes a particular prediction. This is especially crucial in medical settings, where the model must rely on the correct regions of the image for its decisions to be trustworthy.

In this project we will optimise a CNN to classify MRI brain scans into the following categories: no tumour, pituitary tumour, meningioma tumour and glioma tumour. We then explore different interpretation models to help understand what pixels and features the CNN has used in order to classify the image. To assess the performance of these different models we look at the ARoP_MoRF metric as well as visually compare the saliency maps. Finally we will address how these different models scale to a larger, real world setting.
